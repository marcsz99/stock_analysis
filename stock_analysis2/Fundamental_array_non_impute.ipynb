```python
tickers = final_dataset_non_impute.keys()
attr_1 = final_dataset_non_impute['MMM'].index

for ticker in tickers:
    attr_2 = list(final_dataset_non_impute[ticker].index)

    for i in range(len(attr_1)):
        if attr_1[i] != attr_2[i]:
            print(ticker, 'has failed')



```

```python
def get_array_non_impute(final_dataset_non_impute):
    tickers = list(final_dataset_non_impute.keys())
    attributes = list(final_dataset_non_impute[tickers[0]].index)
    num_attributes, num_quarters = final_dataset_non_impute[tickers[0]].shape
    array_non_impute = [np.zeros((len(tickers), num_attributes)) for i in range(num_quarters)]
    
    for i, ticker in enumerate(tickers):
        df_ticker = final_dataset_non_impute[ticker]
        for j in range(num_quarters):
            array_non_impute[j][i, :] = df_ticker.iloc[:, j]
    
    array_non_impute = [pd.DataFrame(arr, columns = attributes, index = tickers) for arr in array_non_impute]
    
    return array_non_impute
```

```python
array_non_impute = get_array_non_impute(final_dataset_non_impute)
```

```python
with open("/Users/marcs/OneDrive/Documents/stock_analysis2/array_non_impute.pickle", 'wb') as f:
        pickle.dump(array_non_impute, f)
```
